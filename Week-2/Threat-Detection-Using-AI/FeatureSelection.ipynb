```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# List of attack types and benign type
attack_types = ["Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", "DoS slowloris", "FTP-Patator",
                "Heartbleed", "Infiltration", "PortScan", "SSH-Patator", "Web Attack  Brute Force",
                "Web Attack  Sql Injection", "Web Attack  XSS"]
benign_type = "BENIGN"


def perform_feature_selection(data):
     # Function for performing feature selection that uses RandomForestRegressor
    X = data.drop(columns=[" Label"])
    y = (data[" Label"] != benign_type).astype(int)
    
    clf = RandomForestRegressor(n_estimators=100, random_state=42)
    clf.fit(X, y)
    
    importances = clf.feature_importances_
    return importances

def plot_feautre_importance(importance_df, attack_type):
    # Plot and save the feautre importance as png
    plt.figure(figsize=(10, 6))
    importance_df.head(10).plot(kind="bar", x="Feature", y="Importance", legend=None)
    plt.title(f"Feature Importance for {attack_type}")
    plt.xlabel("Feature")
    plt.ylabel("Importance")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"{attack_type}_importance_plot.png")
    plt.show()

for attack_type in attack_types:
    # Load the attack vs benign file
    input = f"{attack_type}_vs_{benign_type}.csv"
    attack_data = pd.read_csv(input, low_memory=False)
    
    # Perform feature selection
    importances = perform_feature_selection(attack_data)
    
    # Create a DataFrame for "importance" list
    importance_df = pd.DataFrame({
        "Feature": attack_data.drop(columns=[" Label"]).columns,
        "Importance": importances
        })
    
    # Calculate Percentage
    importance_df["Percentage"] = (importance_df["Importance"] / importance_df["Importance"].sum()) * 100
        
    # Sort the DataFrame by importance in descending order
    importance_df.sort_values(by="Importance", ascending=False, inplace=True)
    
    # Print the top 10 features and their percentages
    print(f"\nTop 10 features and their percentages for {attack_type}:")
    print(importance_df.head(10))
    
    # Save the importance list to a CSV file
    importance_df.to_csv(f"{attack_type}_importance.csv", index=False)
    print(f"Saved importance list for {attack_type} attack")
    
    # Plot feature importance
    plot_feautre_importance(importance_df, attack_type)

print("Feature selection and visualization completed for all attack types.")
```
